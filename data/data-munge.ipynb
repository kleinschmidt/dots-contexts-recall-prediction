```julia
using MAT, DataFrames
```

```julia
dot2014 = matread("./Dot\ 2014/data/original_data.mat")["d"]
```

```julia
varnames = ["subjid1", "block", "rep", "respnr", "line", "rad", "var", "pred", 
            "new_block", "correct", "resp_time", "new_rho_bias", "rho", "theta_resp", 
            "theta", "rho_resp", "rep_number", "new_theta_resp", "new_theta_bias"] 
```

```julia
size.(collect(values(dot2014)))
```

```julia
dat = DataFrame([vec(dot2014[name]) for name in varnames], Symbol.(varnames))
```

```julia
using Underscore, DataFramesMeta
```

```julia
@_ dat |> 
    @where(_, :block .> 1)
```

```julia
pred = @where(dat, :pred .> 0)
```

```julia
recall = @_ dat |> 
    @where(_, :pred .== 0) |> 
    @transform(_ ,
               x = cos.(:theta) .* :rho./2π,
               y = sin.(:theta) .* :rho./2π,
               x_resp = cos.(:theta_resp) .* :rho_resp./2π,
               y_resp = sin.(:theta_resp) .* :rho_resp./2π)
```

```julia
using Plots, StatPlots

@_ recall |> @where(_, :subjid1 .== 1) |> @with(_, plot(:theta, :rho, seriestype=[^(:path), ^(:scatter)], proj=^(:polar)))
```

```julia
@_ recall |> @where(_, :block .== 1.0) |> @with(_, scatter(:x, :y, group=:line))
```

```julia
@with(recall, histogram2d(:theta, :rho))
```

```julia
@with(recall, histogram(:rho))
```

```julia
@with(recall, histogram(:theta, nbins = 100, xticks=[0, π/2, π, 3π/2, 2π]))
```

```julia
@with(recall, scatter(:rad, :rho))
```

```julia
using CSV

CSV.write("dots2014_recall.csv", recall)
CSV.write("dots2014_prediction.csv", pred)
```

```julia
using JLD2

@save "dots2014.jld2" recall pred
```
