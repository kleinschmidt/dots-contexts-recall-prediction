```julia
colnames = ["angle_resp", "angle_resp_pred", "angles_pred", "angles_recall", "correct",
"correct_pred", "pred_types", "rad_resp", "rad_resp_pred", "rads_pred",
"rads_recall", "x_coord", "x_coord_pred", "y_coord", "y_coord_pred", "y_pred",
"x_pred"]
```

```julia
function sep_recall_pred(mat2017)
    recall = DataFrame(vec.(getindex.(mat2017, nonpred_colnames)),
                       Symbol.(replace.(nonpred_colnames, r"_recall$", "")))
    pred = DataFrame(vec.(getindex.(mat2017, pred_colnames)),
                     Symbol.(replace.(pred_colnames, r"_pred$", "")))
    recall[:subnum] = mat2017["subnum"]
    pred[:subnum] = mat2017["subnum"]
    recall, pred
end

r1, p1 = sep_recall_pred(mats2017[1])
r1
```

```julia
function roundto(options, x; dist = (x,y) -> abs(x-y)) 
    x,ind = findmin(dist(x,opt) for opt in options)
    options[ind]
end
```

```julia
#radii = [0, 0.1, 0.3, 0.5, 0.7, 0.9]
radii = [0, 0.7, 0.5, 0.1, 0.9, 0.3, 0.7, 0.1, 0.7, 0.3, 0.9, 0.5, 0.1, 0.9, 0.5]
angles = [0, 0, 8, 45, 82, 90, 116.5, 153.5, 180, 188, 225, 262, 270, 296.5, 333.5]


struct Polar{T}
    ρ::T
    θ::T
end

xy(p::Polar) = [cos(p.θ), sin(p.θ)] .* p.ρ

distance(p1::Polar, p2::Polar) = norm(xy(p1) - xy(p2))

polar_clus = Polar.(radii, angles ./ (360/2π))
polar_recall = Polar.(r1[:rads_recall], r1[:angles_recall] ./ (360/2π))
[findmin(distance.(polar_clus, p))[2] for p in polar_recall]

roundto.((polar_clus, ), Polar.(r1[:rads_recall], r1[:angles_recall]./(360/2π)), dist=distance)
```

```julia
[findmin(distance.(polar_clus, p))[2] for p in Polar.(r1[:rads_recall], deg2rad.(r1[:angles_recall]))]
```

```julia
function determine_cluster!(d)
    polars = Polar.(d[:rads], deg2rad.(d[:angles]))
    idxs = last.([findmin(distance.(polar_clus, p)) for p in polars])
    clusters = polar_clus[idxs]

    d[:cluster] = idxs
    d[:rad_cluster] = getfield.(clusters, :ρ)
    d[:angle_cluster] = rad2deg.(getfield.(clusters, :θ))
   
    d[:newblock] = [1, diff(d[:cluster])...] .!== 0
    d[:block] = cumsum(d[:newblock])
    d
end

determine_cluster!(r1)
```

```julia
determine_cluster!(p1)
```

```julia
rp = sep_recall_pred.(mats2017)
recall_2017 = mapreduce(determine_cluster!, vcat, first.(rp))
pred_2017 = mapreduce(determine_cluster!, vcat, last.(rp))

@show size(recall_2017)
@show size(pred_2017)
```
